## 📚 Basic Tutorial - Protocolo

**1. Procurando Estruturas Similares com a Proteína**
- Resgatar sequencia FASTA da sua proteína no NCBI
- Utilizar o protein BLAST com o database do PDB para resgatar as estruturas mais próximas do template
- Baixar a o arquivo.pdb das 6 estruturas mais próximas
- Construir o arquivo.ali a partir da sequência FASTA
- Modificar o Header: 



- É necessário colocar o * no final da cadeia para indicar o fim.
- Script build_profile.py -> Analisa a sequencia FASTA e procura no banco de dados do PDB as similares
- Necessário modificar o nome do arquivo.ali dentro do script para ser igual ao da sua proteína
- Necessário ter o arquivo pdb_95.pir
- No terminal (entrar como adm) entre no diretório que quer guardar as informações
- mod10.5 build_profile.py
- Output -> build_profile.log
- Mostra informações gerais sobre a proteína e quais estruturas do pdb são similares

**2. Selecionando e Alinhando os Templates**
- Necessário colocar o arquivo.pdb das 6 estruturas similares na pasta
- Vamos utilizar o script compare.py
- Modificar o nome das estruturas presentes nesse script
- Mod10.5 compare.py
- Output -> compare.log
- Mostra um dendrograma de correlação estrutural entre os templates utilizados e a sequencia FASTA



**3. Alinhamento da Estrutura com a Sequência**
- Escolher a melhor estrutura analisada a partir do dendrograma
- Vamos utilizar o script align2d.py
- É necessário modificar os nomes dos arquivos dentro do script
- Mod10.5 align2d.py
- Output -> TvLDH-1bdmA.ali e TvLDH-1bdmA.ali.pap
- Alinha as sequências do arquivo .ali (FASTA) com a estrutura escolhida
- É possível ver nessa etapa quais aminoácidos que estão faltando e quais estão preservados

**4. Construção dos Modelos**
- Construção de 5 modelos a partir do arquivo TvLDH-1bdmA.ali
- Vamos utilizar o script build_model.py
- É necessário modificar os nomes dos arquivos no script
- Output -> Arquivos .pdb dos modelos criados



- O melhor modelo é o que tem o menor valor de DOPE score (energia)
- Nesse caso, é o modelo 2

**5. Avaliação do Modelo**
- Avaliar o potencial DOPE do modelo criado
- Vamos utilizar o script evaluate_model.py
- É necessário modificar os nomes dos arquivos no script
- Output -> arquivo.profile
- Esse arquivo pode ser utilizado para plotar um gráfico de DOPE-score por resíduo 
- Utilizar o script plot_profiles.py para plotar o gráfico
