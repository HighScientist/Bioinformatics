##  Basic Tutorial - Protocolo

**1. Procurando Estruturas Similares com a Protena**
- Resgatar sequencia FASTA da sua protena no NCBI
- Utilizar o protein BLAST com o database do PDB para resgatar as estruturas mais prximas do template
- Baixar a o arquivo.pdb das 6 estruturas mais prximas
- Construir o arquivo.ali a partir da sequncia FASTA
- Modificar o Header: 



-  necessrio colocar o * no final da cadeia para indicar o fim.
- Script build_profile.py -> Analisa a sequencia FASTA e procura no banco de dados do PDB as similares
- Necessrio modificar o nome do arquivo.ali dentro do script para ser igual ao da sua protena
- Necessrio ter o arquivo pdb_95.pir
- No terminal (entrar como adm) entre no diretrio que quer guardar as informaes
- mod10.5 build_profile.py
- Output -> build_profile.log
- Mostra informaes gerais sobre a protena e quais estruturas do pdb so similares

**2. Selecionando e Alinhando os Templates**
- Necessrio colocar o arquivo.pdb das 6 estruturas similares na pasta
- Vamos utilizar o script compare.py
- Modificar o nome das estruturas presentes nesse script
- Mod10.5 compare.py
- Output -> compare.log
- Mostra um dendrograma de correlao estrutural entre os templates utilizados e a sequencia FASTA



**3. Alinhamento da Estrutura com a Sequncia**
- Escolher a melhor estrutura analisada a partir do dendrograma
- Vamos utilizar o script align2d.py
-  necessrio modificar os nomes dos arquivos dentro do script
- Mod10.5 align2d.py
- Output -> TvLDH-1bdmA.ali e TvLDH-1bdmA.ali.pap
- Alinha as sequncias do arquivo .ali (FASTA) com a estrutura escolhida
-  possvel ver nessa etapa quais aminocidos que esto faltando e quais esto preservados

**4. Construo dos Modelos**
- Construo de 5 modelos a partir do arquivo TvLDH-1bdmA.ali
- Vamos utilizar o script build_model.py
-  necessrio modificar os nomes dos arquivos no script
- Output -> Arquivos .pdb dos modelos criados



- O melhor modelo  o que tem o menor valor de DOPE score (energia)
- Nesse caso,  o modelo 2

**5. Avaliao do Modelo**
- Avaliar o potencial DOPE do modelo criado
- Vamos utilizar o script evaluate_model.py
-  necessrio modificar os nomes dos arquivos no script
- Output -> arquivo.profile
- Esse arquivo pode ser utilizado para plotar um grfico de DOPE-score por resduo 
- Utilizar o script plot_profiles.py para plotar o grfico
