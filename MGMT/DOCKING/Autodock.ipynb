## 📚 Autodock Vina - Protocolo

📌 [Tutorial](https://vina.scripps.edu/tutorial/)

**1. AutoDock Tools**
- Vamos obter os arquivos.pdbqt da proteína e do ligante, além das configurações da Caixa
- File -> Read Molecule -> arquivo.pdb da Proteína
- Edit -> Hydrogens -> Add -> Polar Only -> OK
- Edit -> Charges -> Add Kollman Charges
- Grid -> Macromolecule -> Choose -> Protein -> Select Molecule -> Salvar o arquivo.pdbqt
- Grid -> Grid Box -> Definir as coordenadas e as dimensões da Caixa
- Grid Options -> File -> Output grid dimensions file -> Salvar o grid.txt
- Espaçamento (Vina) -> 1 Angstrom
- Espaçamento (AutoDock-4) -> 0,375 Angstrom
- Salvar as informações da Caixa para utilizar no Vina posteriormente
- Ligand -> Input -> Open -> arquivo.pdb do Ligante
- Ligand -> Torsion Tree -> Choose Torsions -> Dá para modificar as ligações rotacionáveis
- Ligand -> Output -> Save as PDBQT


**2. AutoDock Vina**
- Agora que temos os arquivos.pdbqt da proteína e do ligante, vamos produzir um arquivo de configurações
- Abra um arquivo.txt (config.txt) e digite as configurações:

- receptor = protein.pdbqt
- ligand = ligand.pdbqt

- all = all.pdbqt

- center_x = 11
- center_y = 90.5
- center_z = 57.5

- size_x = 15
- size_y = 15
- size_z = 15

- energy_range = 4

- exhaustiveness = 8

- No prompt entre na pasta que estão os 3 arquivos e coloque as configurações:
- vina1.1.2 --receptor protein.pdbqt --ligand ligand.pdbqt --config config.txt --log log.txt --out output.pdbqt

- Obs: É possível colocar todos os parâmetros dentro do arquivo de config, depois só rodar vina1.1.2 --config config.txt
